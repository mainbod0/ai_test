```python
import pandas as pd

df = pd.read_csv('../comp/weather/Summary of Weather.csv', parse_dates=['Date'])[['Date', 'MeanTemp', 'STA']]
df = df[df['STA'] == 10001]
df = df.drop('STA', axis=1)
```

```python
df.columns = ['rdate', 'temp']
df = df.set_index('rdate')
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,4))
plt.plot(df)
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

res = seasonal_decompose(df['temp'], model='additive', period=12)
plt.figure(figsize=(12,4))
plt.plot()
plt.show()
```
